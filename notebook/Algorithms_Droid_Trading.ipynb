````python
# -*- coding: gbk -*-
import os
import sys
from win32api import GetSystemMetrics

class TradingClient:
    """ Initializing the new Trading client ,Click set windows control order
    ,then windows Control Names setted , We can operate the Client by WinControls """
    def __init__(self,debug=0):
        self.debug = debug      # if in debugging ,assumed WinControls in Order
        self.ordered = False if debug==0 else True
        self._WinLocked = "交易界面已锁定"
        self._mainWinTitle = "中投证券超强版独立委托"
        self._listview_cmd = "GetText"
        self._holdlist=["SysListView326", (0, 3, 4, 5, 8)]
        self._deallist = ["SysListView322", (1, 2, 4, 7, 8, 9)]  # add 1 for bid code
        self._historylist = ["SysListView321", (0, 2, 3, 5, 8, 9, 10)] # add  2 bid code
        self.autoit = self.getAutoitComControl()
        self.checkClientRuning()
        # Find Open Client and Activated it and Focus on it.

    def checkClientRuning(self):
        """see if the proxy client is running now!"""
        try:
            assert self.FindActive() == True
            self.unLock()
        except AssertionError:
            print " <<==== Client Not Running!! Exit===>"
            exit(1)
        if not self.debug:
            self.ordered = self.SetWinOrder()

    def getAutoitComControl(self):
        """ In Case of Successful Operating the Trading Client
        the Automatically Trading machine must be 32 bit of every thing
        from Windows to autoit and python, In X64 Windows Some WinControls may
        be dysfunctional"""
        try:
            import win32com.client
        except ImportError:
            raise ImportError, 'This program requires the pywin32 extensions for Python. See http://starship.python.net/crew/mhammond/win32/'

        import pywintypes  # to handle COM errors.
        # Import AutoIT (first try)
        try:
            _autoit = win32com.client.Dispatch("AutoItX3.Control")
        except pywintypes.com_error:
            print "com_error of pywintypes , try to register COM control again"
            # If can't instanciate, try to register COM control again:
            os.system("regsvr32 /s AutoItX3.dll")
        # Import AutoIT (second try if necessary)
        if not _autoit:
            try:
                _autoit = win32com.client.Dispatch("AutoItX3.Control")
            except pywintypes.com_error:
                raise ImportError, "Could not instanciate AutoIT COM module because"

        if not _autoit:
            print "Could not instanciate AutoIT COM module."
            sys.exit(1)
        return _autoit


    def FindActive(self):
        """ Let AutoIT find the windows and focus to it,
        notice the encoding may be an Error, currently set to gbk
        print self._mainWinTitle  """
        self.autoit.AutoItSetOption("WinTitleMatchMode", 4)
        hwin = self.autoit.WinGetHandle(self._mainWinTitle)
        if  hwin == u'0x0000000000000000':
            print "window handle not exist by win Title"
            return False
        else:
            hwin = "[HANDLE:%s]" % hwin
            if not self.autoit.WinActive(hwin):
                self.autoit.WinActivate(hwin, "")
            self.autoit.WinSetState("", "", self.autoit.SW_MAXIMIZE)  #maxsize the windows
            self.autoit.Sleep(1200)
            return True

    def unLock(self):
        """if Successfully unlock and maximizing the windows return 1 else 0"""
        import base64
        pw = base64.decodestring('MzExODE1\n') # simply enconding the password
        self.autoit.ControlSetText(self._mainWinTitle,self._WinLocked, "AfxWnd421", pw)
        self.autoit.ControlClick("", "", "Button2")
        self.autoit.Sleep(1200)
        res =self.autoit.WinSetState("", "", self.autoit.SW_MAXIMIZE)  #maxsize the windows

        self.autoit.Sleep(1200)
        return res

    def getListViewByPos(self, ctrlname, cmd, columnlist):
        """20160316 Decision : Get Listview by in Rows
        input cltrname , cltr_cmd , and interested cloumnlist
        return : Rows of cloumnlist
        Usage : Called different functions to get specific List"""
        fLine = []
        Rows = []
        row = 0
        while  self.autoit.ControlListView(self._mainWinTitle, "", ctrlname, cmd, row, 0):
            for i in columnlist:  # stockcode,available,cost,nowprice, gainLost
                fi = self.autoit.ControlListView(self._mainWinTitle, "", ctrlname, cmd, row, i)
                fLine.append(fi)
            Rows.append(fLine)
            # print "Listview Items 1 Line: ===> ",fLine
            fLine = []
            row += 1
        return Rows

    def getHoldingList(self):
        self.MouseClickPosbyName("hands")
        ctlrname = self._holdlist[0]
        col_list = self._holdlist[1]
        rows = self.getListViewByPos(ctlrname, self._listview_cmd  ,col_list)
        return self.getFormattedHoldingListRows(rows)

    def getDealMadeList(self):
        """if not click the button ,should SysListView content updated
        !!! tricky thing is the order of clicking the button can change the
        !!! Control SysListView32  last number !!!"""
        self.MouseClickPosbyName("deal")
        ctlrname, col_list = self._deallist[0],self._deallist[1]
        rows = self.getListViewByPos(ctlrname, self._listview_cmd  ,col_list)
        # print "returned Deal Done rows>>>", rows, "<<<<"
        return self.getFormattedDealMadeListRows(rows)

    def getHisTradeList(self):
        """ read from trade client, get the history trade infomation"""
        self.MouseClickPosbyName("history")
        ctlrname, col_list = self._historylist[0],self._historylist[1]
        rows = self.getListViewByPos(ctlrname, self._listview_cmd  ,col_list)
        # print "returned Deal Done rows>>>", rows, "<<<<"
        return self.getFormattedHistoryTradeList(rows)

    # @getListViewByPos( "SysListView321", "GetText",[0, 3, 4, 5, 8])
    def getFormattedHoldingListRows(self, Rows):
        """formatting the Rows with corresponding types and values"""
        if not Rows:
            return []
        rows  = []
        for fLine in Rows:
            fLine[0] = fLine[0].encode('ascii')
            fLine[1:] = map(float, fLine[1:])
            rows.append(fLine)
        return rows

    #[['002008', 1700.0, 0.609, 10.27, 16424.06],
    #['300122', 2100.0, 36.238, 35.81, -899.16],
    #['300177', 100.0, 6.233, 15.45, 921.69]]

    def getFormattedDealMadeListRows(self, Rows):
        """get formatting the Rows with corresponding types and values"""
        if not Rows:
            return []
        rows  = []
        for fLine in Rows:
            print "Should be bidding Code==>", fLine[0]
            fLine[3:] = map(float, fLine[3:])
            fLine[0] = fLine[0].encode('ascii')  # bid code
            fLine[1] = fLine[1].encode('ascii')
            fLine[2] = fLine[2].encode('gbk')
            fLine[2] = -1 if fLine[2] == '买入' else 1
            rows.append(fLine)
        return rows

    def getFormattedHistoryTradeList(self, Rows):
        """get formatting the Rows with corresponding types and values"""
        rows = []
        if not Rows:
            return []
        for fLine in Rows:
            fLine[4:] = map(float, fLine[4:])
            fLine[0] = fLine[0].encode('ascii')  # date
            fLine[1] = fLine[1].encode('ascii')  # bidding code
            fLine[2] = fLine[2].encode('ascii')  # stock code
            fLine[3] = fLine[3].encode('gbk')
            fLine[3] = -1 if fLine[3] == '买入' else 1
            rows.append(fLine)  #MemoryError here
        return rows


    def CancelOrder(self):
        """click the cancel button and get the order info from the SysListView324
        Todo ====>>> """
        pass


    def lastInHandTrade(self, numberofdays=1):
        """when first called ,return the last trading date deal made,
        but when today commited deal just return todays' trades"""
        Rows = getHisTradeList()
        InHand_lst = getHoldingList()
        stocks_inHands = getListCol(InHand_lst, 0)
        #checkPointList = ParseTrade(Rows,stocksInfoList)
        checkPointList = getPreviousTrades(Rows, numberofdays)
        #print "latest date trades==== >", checkPointList,len(checkPointList)
        lastTradelst = [row[1:] for row in checkPointList if row[1] in stocks_inHands]
        #print "in Hand stocks list=====>",lastTradelst,len(lastTradelst)
        return lastTradelst


    def getPreviousTrades(self, Rows, numberdays=2):
        DateList = list(set([row[0] for row in Rows]))
        DateList.sort()
        #lastDate = DateList[-2:]  # last record first is the latest trading date
        lastDate_lst = DateList[-1 * numberdays:]
        #print "Last Date Array ===> ", lastDate_lst
        #lastTradelst = [row for row in Rows if row[0] == lastDate]
        lastTradelst = [row for row in Rows if row[0] in lastDate_lst]
        return lastTradelst


    def retrieveNum(self, crazystring):
        """retrive digital number and . from mixed string """
        floatstring = filter(lambda ch: ch in '1234567890.', crazystring)
        return floatstring


    def getGeneralInfo(self):
        """余额:7869.99  可用:7869.99  可取:7869.99  参考市值:94205.00
        资产:102074.99  盈亏:16446.59   """
        Ginfo = self.autoit.ControlGetText(self._mainWinTitle, "", "Static226")
        Ginfo = Ginfo.encode('gbk')
        print "General Account info ==>%s<<<"%Ginfo
        assert len(Ginfo) > 1
        strList = Ginfo.split('  ')[:-1]  # filter out the last blank string
        accGenList = map(self.retrieveNum, strList)
        accGenList = map(float, accGenList)
        print ">>>>%s<<<<"%accGenList
        return accGenList

    def getResolution(self):
        """  find the current screen resolution
        print "width =", GetSystemMetrics(0)
        print "height =",GetSystemMetrics(1)
        @return:
        """
        return GetSystemMetrics(0), GetSystemMetrics(1)

    def MouseClickPosbyName(self, name):
        """
        @param name:name the button name
        @return: click the correspondate button pos x,y
        x1/width1 = x0/width0  y1/height1 = y0/height0
        """
        #redefine the positon in 1280,800
        dict = {"buy": (234, 38),  #for the 1024X768 convert resolution 800X600
         "sell": (270, 38),        # 267/1024=x1/800 36/768=y1/600
         "cancel": (310, 38),
         "deal": (350, 38),
         "hands": (390, 38),
         "refresh": (430, 38),
         "history": (70, 272)}
        if name in dict.keys():
            x, y = dict[name]
            print "Name == %s to click the button pos %s "%(name, dict[name])
            self.autoit.MouseClick("left",x,y,1,40)
            self.autoit.Sleep(1000)


    #Control Name changes when click in different sequence!!!
    #Below use ID instead!!
    def fillOrderData(self, type, code, price, hands):
        if type == "buy":
            self.MouseClickPosbyName("buy")
            btntext = "买入下单",
            win_str = "买入交易确认"
            confirm_btn_str = "买入确认",
            # Ctl_lst = ["Edit32", "Edit33", "Edit35", "Button134"]
            Ctl_lst = ["Edit15", "Edit16", "Edit18", "Button73"]
        elif type == "sell":
            self.MouseClickPosbyName("sell")
            btntext = "卖出下单",
            win_str = "卖出交易确认"
            confirm_btn_str = "卖出确认"
            #winsound.Beep(100, 300)
            # Ctl_lst = ["Edit21", "Edit22", "Edit24", "Button125"]
            Ctl_lst = ["Edit27", "Edit28", "Edit30", "Button85"]
        # print "Code=>%s Price=>%s Hands ==> %s"%(code,price,hands)
        # print "Control List ==> %s"%Ctl_lst
        self.autoit.ControlSetText(self._mainWinTitle, "", Ctl_lst[0], code)  # stockcode )
        self.autoit.ControlSetText(self._mainWinTitle, "", Ctl_lst[1], str(price))  # "3.11")
        self.autoit.ControlSetText(self._mainWinTitle, "", Ctl_lst[2], str(hands))  # "200")
        #self.autoit.Send(code+str(price)+"{Tab}"+str(hands))
        self.autoit.Sleep(2000)
        #self.autoit.ControlClick("",btntext,Ctl_lst[3])
        if type == "sell":
            self.autoit.MouseClick("left", 368, 212)
        elif type == "buy":  # type = "buy"
            self.autoit.MouseClick("left", 368, 232)
        #self.autoit.Send("{ENTER}")
        self.autoit.WinWait(win_str)
        #self.autoit.ControlClick(win_str,confirm_btn_str,"Button1")
        self.autoit.Send("{ENTER}")
        self.autoit.WinWait("提示")
        self.autoit.Sleep(5000)
        #self.autoit.Send("{ENTER}")  # the last confirm
        self.autoit.MouseClick("left", 744, 447)
        self.autoit.Sleep(5000)

    def SetWinOrder(self):
        """
            @return: the proper order of creating windows widget
        """
        for name in ["history","deal","buy","sell","cancel","hands","refresh"]:
            print "the name====>" ,name
            self.MouseClickPosbyName(name)
            self.autoit.Sleep(1000)
        return True


def ArrayPrint(theArray):
    """ print each line of an Array
    """
    for item in theArray:
        print item



if __name__ == '__main__':
    proxy = TradingClient(debug=1)
    #proxy.CheckEnvironment()
    #getDailyDatum()
    proxy.FindActive()
    print "Screen Size ===> ", proxy.getResolution()
    if proxy.autoit.WinActive(proxy._mainWinTitle, proxy._WinLocked):
        proxy.unLock()
    proxy.SetWinOrder()
    # proxy.MouseClickPosbyName("deal")
    print "Now in Hands stocks===> \n "
````

````python
# 需要引入的Python 库
import numpy as np
import time
import pandas as pd
import datetime
from urllib import urlopen
````

````python
def getStockData(stockCode):
    """At the Market time use urllib to open and get real time stock
    biddin gand pricing ,delay exist ,alway keep alarm to it
     eg：http://hq.sinajs.cn/list=sz002594 """
    HqString = "http://hq.sinajs.cn/list=" + stockCode
    # from the url return data, retrive the data items ,return the price List
    # reading realtime bidding info from sina
    while True:  # if IOError wait 30seconds to retry
        try:
            begin_t = time.time()
            hqList = urlopen(HqString).read()
            consume_t = time.time() - begin_t
            # print "getting from remote hq.sina,consumes %f seconds....." % consume_t
            # logger.info("Check if the HqString changed!!==> \n %s" % hqList)
            break
        except IOError:
            print "IOError ,sleep 20 second,then fetch again"
            time.sleep(20)
    hqList = hqList.split(',')
    # print "====retrieved Hq from sina ,the length ===>", len(hqList)
    if len(hqList) != 33:
        print "Length Error != 33 Hqlist is invalid!!!!!!!!!!! return 0  \n"
        print "Error List contains===>", hqList
        return 0
    todayOpen, yesClose, atTime = hqList[1], hqList[2], hqList[31].split('"')[0]
    # tmpHigh,tmpLow, tmpVol ,tmpMoney =  hqList[4], hqList[5], hqList[8] , hqList[9]
    # nowPrice = hqList[3]
    valList = map(float, hqList[1:30]) + [hqList[30] + " " + atTime]
    return valList
````
